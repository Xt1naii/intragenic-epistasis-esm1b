# Общий датасет

```python
data = pd.concat([ubc9, yap1, pabp, ube4b, brca1_db6, brca1_db8, spg1], ignore_index=True)
```

```python
data.shape
```

```python
data = data.dropna()
```

```python
data = data[data["Score.A'"] != -np.inf]
```

```python
data.isna().any()
```

```python
data.shape
```

Score.Delta = Epsilon = Sac - Sa - Sc = Sa' - Sa

```python
data['Score.Delta.Esm1b'] = data["Score.A'.Esm1b"] - data["Score.A.Esm1b"]
```

```python
data
```

```python
# spearmanr
corr_sa_sa_sp = pd.DataFrame(columns=[data.Protein.unique()], index=['spearmanr', 'p_value'])

for prot in data.Protein.unique():
    corr_res, p_value = spearmanr(data[data['Protein'] == prot]["Score.A'"], data[data['Protein'] == prot]["Score.A'.Esm1b"])
    corr_sa_sa_sp.loc['spearmanr', prot] = corr_res
    corr_sa_sa_sp.loc['p_value', prot] = p_value

corr_sa_sa_sp
```

```python
corr_delta_delta_sp = pd.DataFrame(columns=[data.Protein.unique()], index=['spearmanr', 'p_value'])

for prot in data.Protein.unique():
    corr_res, p_value = spearmanr(data[data['Protein'] == prot]["Score.Delta"], data[data['Protein'] == prot]["Score.Delta.Esm1b"])
    corr_delta_delta_sp.loc['spearmanr', prot] = corr_res
    corr_delta_delta_sp.loc['p_value', prot] = p_value

corr_delta_delta_sp
```

### UBC9_HUMAN

```python
shifts[(shifts.Protein == 'UBC9_HUMAN') & (shifts['Counts.E'] >= 5)].sort_values('Pos.C')
```

```python
def std(x): 
    return np.std(x, ddof=0)
```

```python
ubc9_shift = data[data.Protein == 'UBC9_HUMAN'].groupby(['Pos.C', 'AA1.C', 'AA2.C'], as_index=False).agg({'Score.C.Esm1b': 'mean', 'Score.Delta': ['mean', std], 'Score.Delta.Esm1b': ['mean', std], 'Protein': 'count'})
ubc9_shift.columns = ubc9_shift.columns.map('.'.join).str.strip('.')
ubc9_shift = ubc9_shift[ubc9_shift['Protein.count'] >= 5].round(3)
```

```python
ubc9_shift
```

```python
shifts[(shifts.Protein == 'UBC9_HUMAN') & (shifts['Counts.E'] >= 5)].sort_values('Pos.C')
```

```python
print('corr_epsilon_mean:', spearmanr(ubc9_shift["Score.Delta.mean"], ubc9_shift['Score.Delta.Esm1b.mean']))
print('corr_epsilon_std:', spearmanr(ubc9_shift["Score.Delta.std"], ubc9_shift['Score.Delta.Esm1b.std']))
```
